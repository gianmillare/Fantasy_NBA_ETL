```python
from bs4 import BeautifulSoup as bs
from splinter import Browser as browser
import pandas as pd
import time
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_players_in_the_Naismith_Memorial_Basketball_Hall_of_Fame'
hof_table = pd.read_html(url)
hof_df = hof_table[0]
hof_df = hof_df.loc[hof_df["Year"] > 1987].drop(columns=['Ref.'])

hof = hof_df["Inductees"].to_list()
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)

url = 'https://www.basketball-reference.com/'
browser.visit(url)
```

```python
for x in hof:
    
    html = browser.html
    soup = bs(html, 'html.parser')
    browser.find_by_css('.ac-input').first.fill(x)
    browser.find_by_value('Search').first.click()
#     totals = browser.find_by_id("all_totals")
#     csv = totals.first.find_by_css('.hasmore').click()
    
    
    stats = soup.find_all(id="all_totals")
    time.sleep(1)
    for stat in stats:
        print(stat.get_text())
Collapse



```
