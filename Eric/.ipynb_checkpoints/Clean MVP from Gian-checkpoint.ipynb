```python
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect

import pandas as pd
import numpy as np
```

```python
file_path = "../Gian/Outputs/MVP_stats.csv"
df = pd.read_csv(file_path)
df = df.copy()
df.head(20)
# len(df) 231
```

```python
df = df[df["Season"].notnull()]
df = df[df["Season"].str.contains('-')]
df = df[~df["Season"].str.contains('-20')]
# print(len(df)) 192
df.head(20)
```

```python
engine = create_engine("sqlite:///NBA.sqlite")

df.to_sql(name='MVPs', con=engine, if_exists='replace', index=False)
```

```python
pd.read_sql_query('SELECT * FROM MVPs', con=engine).head()
```
